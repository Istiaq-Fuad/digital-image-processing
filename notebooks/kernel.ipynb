```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("deer.jpg", 0) 
```

```python
image.shape
```

```python
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
kernels = [
    np.array([[ 0, -1,  0],
              [ 0,  1,  0],
              [ 0, -1, 0]]),

    np.array([[ 1,  0, 2],
              [ 1,  2, 1],
              [ 2, -2, 0]]),
 
    np.array([[ 0,  2, -1],
              [-2, -1,  2],
              [ 1,  0, -2]]),

    np.array([[ 2, -2,  1],
              [ 0,  1, -1],
              [-1,  2,  0]]),

    np.array([[-1,  0,  2],
              [ 2, -1, -2],
              [ 0,  1,  1]]),

    np.array([[ 1, -2,  0],
              [ 0,  2, -1],
              [-1,  1, -2]]),

    np.array([[ 0,  1, -1],
              [ 2, -2,  0],
              [-2,  1,  2]]),

    np.array([[ 2, -1,  2],
              [-2,  0, -2],
              [ 1,  1, -1]]),

    np.array([[ 1, -1,  0],
              [ 0,  2, -2],
              [-1,  0,  1]]),

    np.array([[ 0,  2, -2],
              [ 1, -1,  0],
              [-2,  1,  2]]),

    np.array([[ 2,  0, -1],
              [-1, -2,  2],
              [ 0,  1, -2]]),

    np.array([[ 0, -2,  2],
              [ 1,  1, -1],
              [-2,  0,  1]]),

    np.array([[ 1, -1,  2],
              [-2,  0, -1],
              [ 2,  1, -2]]),

    np.array([[ 0,  1, -2],
              [ 2, -1,  0],
              [-1,  2, -2]]),

    np.array([[ 2, -2,  0],
              [ 1,  0, -1],
              [ 0, -1,  2]]),

    np.array([[ 1,  2, -1],
              [ 0, -2,  1],
              [-2,  0,  2]]),

    np.array([[ 0, -1,  2],
              [ 2,  1, -2],
              [-1,  0, -2]]),

    np.array([[ 2,  1, -1],
              [-2, -1,  0],
              [ 1,  0,  2]]),

    np.array([[ 0, -2,  1],
              [ 2,  0, -1],
              [-1,  2, -2]]),

    np.array([[ 1,  0, -1],
              [-2,  2,  0],
              [ 2, -1, -2]]),
]
```

```python
filtered_images = [cv2.filter2D(image, -1, k) for k in kernels]
```

```python
fig, axes = plt.subplots(4, 5, figsize=(15, 12))
axes = axes.ravel()

for i in range(20):
    axes[i].imshow(filtered_images[i])
    axes[i].set_title(f"Kernel {i+1}")
    axes[i].axis("off")

plt.tight_layout()
plt.show()
```

```python
for i, k in enumerate(kernels, start=1):
    print(f"Kernel {i}:\n{k}\n")
```
