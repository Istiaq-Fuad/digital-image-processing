```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def func1(c, r):
    return c * np.log2(1 + r)

def func2(c, r, g):
    return c * r ** g
```

```python
img = cv2.imread('test.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img.shape
```

```python
img_func1 = np.zeros_like(img)
img_func2 = np.zeros_like(img)
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img_func1[i, j] = func1(1, img[i, j])
```

```python
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        img_func2[i, j] = func2(1, img[i, j], 0.8)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(img_func1, cmap='gray')
plt.title('c * np.log2(1 + r)')

plt.subplot(1, 3, 3)
plt.imshow(img_func2, cmap='gray')
plt.title('c * r ** g')

plt.show()
```
