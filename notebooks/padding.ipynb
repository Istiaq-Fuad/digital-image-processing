```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test.jpg')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img_rgb.shape
```

```python
def add_padding(image, top, bottom, left, right):
    height, width, channels = image.shape
    new_height = height + top + bottom
    new_width = width + left + right
    
    padded_image = np.zeros((new_height, new_width, channels), dtype=np.uint8)
    
    padded_image[top:top+height, left:left+width] = image
    
    return padded_image

padded_img = add_padding(img_rgb, 50, 50, 100, 100)
```

```python
plt.imshow(padded_img)
plt.show()
```
